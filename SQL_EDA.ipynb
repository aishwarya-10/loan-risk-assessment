# 3. SQL EDA

```python
connection = sqlite3.connect("loan_application.db")
cur = connection.cursor()
```

## 3.1 Credit Types: 

```python
# Identify the different types of credit offered by the bank.
query1 = """SELECT DISTINCT NAME_CONTRACT_TYPE
            FROM Application
        """
cur.execute(query1)
result1 = cur.fetchall()
columns = ["Credit_types"]
df1 = pd.DataFrame(result1, columns= columns).reset_index(drop=True)
df1
```

## 3.2 Income Distribution & Descriptive Statistics wrt. Credit Type: 

```python
# Analyze income distribution and statistics for each credit type.
# Define the Query
query2 = """SELECT NAME_CONTRACT_TYPE AS credit_type,
                    COUNT(*) AS count,
                    MIN(AMT_INCOME_TOTAL) AS min_income,
                    MAX(AMT_INCOME_TOTAL) AS max_income,
                    AVG(AMT_INCOME_TOTAL) AS mean_income
            FROM Application
            GROUP BY NAME_CONTRACT_TYPE
         """

# Execute the query
cur.execute(query2)

# Fetch results and create DataFrame
result2 = cur.fetchall()
columns = ["Credit_type", "Count", "Min_income", "Max_income", "Mean_income"]
df2 = pd.DataFrame(result2, columns=columns)
df2
```

## 3.3 Analysis of Goods Amount for Cash Loans: 

```python
# Analyze the loan amounts for clients applying for cash loans.
query3 = """SELECT
                (SELECT AMT_GOODS_PRICE
                FROM (
                SELECT *, ROW_NUMBER() OVER (ORDER BY AMT_GOODS_PRICE) AS rn
                FROM Application
                WHERE NAME_CONTRACT_TYPE = 'Cash loans'
                ) AS ranked_data
                WHERE rn = (SELECT COUNT(*) FROM Application WHERE NAME_CONTRACT_TYPE = 'Cash loans') / 2 + 1) AS median_goods_price,
                AVG(AMT_GOODS_PRICE) as avg_goods_price,
                MIN(AMT_GOODS_PRICE) as min_goods_price, 
                MAX(AMT_GOODS_PRICE) as max_goods_price
        FROM Application
        WHERE NAME_CONTRACT_TYPE = 'Cash loans'
        """
cur.execute(query3)
result3 = cur.fetchall()
columns = ["median_goods_price", "avg_goods_price", "min_goods_price", "max_goods_price"]
df3 = pd.DataFrame(result3, columns = columns).reset_index(drop=True)
df3
```

## 3.4 Age Brackets of the Clients: 

```python
# Explore the distribution of clients across different age groups.
query4 = """SELECT 
                CASE
                    WHEN DAYS_BIRTH/365*(-1) BETWEEN 0 AND 17 THEN '0-17'
                    WHEN DAYS_BIRTH/365*(-1) BETWEEN 18 AND 24 THEN '18-24'
                    WHEN DAYS_BIRTH/365*(-1) BETWEEN 25 AND 34 THEN '25-34'
                    WHEN DAYS_BIRTH/365*(-1) BETWEEN 35 AND 44 THEN '35-44'
                    WHEN DAYS_BIRTH/365*(-1) BETWEEN 45 AND 54 THEN '45-54'
                    WHEN DAYS_BIRTH/365*(-1) BETWEEN 55 AND 64 THEN '55-64'
                    ELSE '65+'
                END AS age_bracket,
                COUNT(*) AS Count
            FROM Application
            GROUP BY age_bracket
            ORDER BY age_bracket
            """
cur.execute(query4)
result4 = cur.fetchall()
columns = ["Age Bracket", "Count"]
df4 = pd.DataFrame(result4, columns = columns).reset_index(drop=True)
df4
```

## 3.5 Documents Submission Analysis: 

```python
# Analyze the frequency of document submission by clients.
# Define query
query5 = """SELECT document_name, COUNT(*) AS Freq_submission
            FROM (
                SELECT
                    CASE 
                        WHEN FLAG_DOCUMENT_2 = 1 THEN 'Document 2'
                        WHEN FLAG_DOCUMENT_3 = 1 THEN 'Document 3'
                        WHEN FLAG_DOCUMENT_4 = 1 THEN 'Document 4'
                        WHEN FLAG_DOCUMENT_5 = 1 THEN 'Document 5'
                        WHEN FLAG_DOCUMENT_6 = 1 THEN 'Document 6'
                        WHEN FLAG_DOCUMENT_7 = 1 THEN 'Document 7'
                        WHEN FLAG_DOCUMENT_8 = 1 THEN 'Document 8'
                        WHEN FLAG_DOCUMENT_9 = 1 THEN 'Document 9'
                        WHEN FLAG_DOCUMENT_10 = 1 THEN 'Document 10'
                        WHEN FLAG_DOCUMENT_11 = 1 THEN 'Document 11'
                        WHEN FLAG_DOCUMENT_12 = 1 THEN 'Document 12'
                        WHEN FLAG_DOCUMENT_13 = 1 THEN 'Document 13'
                        WHEN FLAG_DOCUMENT_14 = 1 THEN 'Document 14'
                        WHEN FLAG_DOCUMENT_15 = 1 THEN 'Document 15'
                        WHEN FLAG_DOCUMENT_16 = 1 THEN 'Document 16'
                        WHEN FLAG_DOCUMENT_17 = 1 THEN 'Document 17'
                        WHEN FLAG_DOCUMENT_18 = 1 THEN 'Document 18'
                        WHEN FLAG_DOCUMENT_19 = 1 THEN 'Document 19'
                        WHEN FLAG_DOCUMENT_20 = 1 THEN 'Document 20'
                        WHEN FLAG_DOCUMENT_21 = 1 THEN 'Document 21'
                    END AS document_name
                FROM Application
            ) AS doc_flags
            GROUP BY document_name
            ORDER BY Freq_submission DESC
"""

# Execute the query
cur.execute(query5)

# Fetch results and create DataFrame
result5 = cur.fetchall()
columns = ["document_name", "num_submission"]
df5 = pd.DataFrame(result5, columns=columns)
df5
```

## 3.6 Overall Analysis of Credit Enquiries on Clients: 

```python
# Analyze the number of credit inquiries made by clients before loan application.
# Define query
query6 = """SELECT SUM(AMT_REQ_CREDIT_BUREAU_HOUR) AS inquires_hour,
                    SUM(AMT_REQ_CREDIT_BUREAU_DAY) AS inquires_day,
                    SUM(AMT_REQ_CREDIT_BUREAU_WEEK) AS inquires_week,
                    SUM(AMT_REQ_CREDIT_BUREAU_MON) AS inquires_month,
                    SUM(AMT_REQ_CREDIT_BUREAU_QRT) AS inquires_quarter,
                    SUM(AMT_REQ_CREDIT_BUREAU_YEAR) AS inquires_year
            FROM Application
        """

# Execute the query
cur.execute(query6)

# Fetch results and create DataFrame
result6 = cur.fetchall()
columns = ["inquiries_hour", "inquiries_day", "inquiries_week", "inquiries_month", "inquiries_quarter", "inquiries_year"]
df6 = pd.DataFrame(result6, columns=columns)
df6
```

## 3.9 Integration of previous application data: 

```python
# Integrate insights from the previous_application.csv data to enhance risk assessment.

# Load Data
df_Prev_application = pd.read_csv("Data/previous_application.csv")

# Insert DF to SQL server
df_Prev_application.to_sql("Prev_Application", connection, if_exists="replace") 
```

```python
# Define the Query
query9 = """WITH PreviousApplicant AS (
                SELECT
                    SK_ID_CURR AS Client_ID,
                    AMT_CREDIT, 
                    CASE WHEN NAME_CONTRACT_STATUS = 'Approved' THEN 1 ELSE 0 END AS STATUS,
                    DAYS_DECISION
                FROM Prev_Application
            ),
            RiskScores AS (
                SELECT
                    C.SK_ID_CURR AS Client_ID,
                    (C.AMT_INCOME_TOTAL * 0.5) +
                    (C.CNT_CHILDREN * -0.1) +
                    (C.DAYS_BIRTH * 0.2) +
                    (P.STATUS * 0.1) +
                    (P.AMT_CREDIT * 0.2) +
                    (P.DAYS_DECISION * -0.05) AS risk_score
                FROM Application AS C
                JOIN PreviousApplicant AS P
                ON C.SK_ID_CURR = P.Client_ID
            ),
            RankedRiskScores AS (
                SELECT
                    Client_ID,
                    risk_score,
                    RANK() OVER (ORDER BY risk_score DESC) AS score_rank,
                    COUNT(*) OVER () AS total_records
                FROM RiskScores
            )
            SELECT
                Client_ID,
                risk_score,
                ((score_rank - 1) * 100.0 / (total_records - 1)) AS risk_percentile
            FROM RankedRiskScores
            ORDER BY risk_score DESC
            LIMIT 15
            """

# Execute the query
cur.execute(query9)

# Fetch results and create DataFrame
result9 = cur.fetchall()
columns = ["Client_ID", "risk_score", "risk_percentile"]
df9 = pd.DataFrame(result9, columns=columns)
df9
```

## 3.10 Top 15 customers and contact reach: 

```python
# Identify the top 15 clients and analyze their contact reach.
# Define the Query
query10 = """WITH RankedScores AS (
                SELECT
                    SK_ID_CURR AS Client_ID,
                    (AMT_INCOME_TOTAL * 0.5) + (CNT_CHILDREN * -0.1) + (-DAYS_BIRTH * -0.2) AS risk_score,
                    RANK() OVER (ORDER BY (AMT_INCOME_TOTAL * 0.5) + (CNT_CHILDREN * -0.1) + (DAYS_BIRTH * 0.2) DESC) AS score_rank,
                    COUNT(*) OVER () AS total_scores
                FROM Application
            )
            SELECT
                Client_ID,
                ((score_rank - 1) * 100.0 / (total_scores - 1)) AS risk_percentile
            FROM RankedScores
            ORDER BY risk_percentile ASC
            LIMIT 15
            """

# Execute the query
cur.execute(query10)

# Fetch results and create DataFrame
result10 = cur.fetchall()
columns = ["Client_ID", "risk_score"]
df10 = pd.DataFrame(result10, columns=columns)
df10
```
